```python
# Импортируем библиотеку Pandas
import pandas as pd

# Импортируем библиотеку NumPy
import numpy as np

import datetime as dt
```

```python
# Подгружаем данные и выводим на экран

installs = pd.read_csv('les8_installs.csv', index_col=0)

logins = pd.read_csv('les8_logins.csv', index_col=0)

display(installs)

display(logins)
```

```python
#Проверим тип данных, с которыми предстоит работать

installs.dtypes
```

```python
logins.dtypes
```

```python
# Делаем дату форматом даты

installs['install_date'] = pd.to_datetime(installs['install_date'], format='%Y-%m-%d')
installs.dtypes
```

```python
logins['event_timestamp'] = pd.to_datetime(logins['event_timestamp'], format='%Y-%m-%d')
logins.dtypes
```

```python
logins
```

```python
# Сделаем проверку на дубликаты

installs.duplicated().sum()
```

```python
logins.duplicated().sum()
```

```python
logins = logins.drop_duplicates()
logins
```

```python
# Создадим когорты по дате установки

installs['week_start'] = installs['install_date'].dt.to_period('W').dt.start_time
installs
```

```python
# Смержим таблицы, для того,
# чтобы у каждого пользователя в каждый момент логина была известна дата установки и его отношение к когорте

logins = logins.merge(
    installs,
    on='client_id',
    how='inner'
)
logins
```

```python
# Рассчитаем метрику LT

logins['lifetime'] = logins['event_timestamp'] - logins['install_date']
logins['lifetime'] = logins['lifetime'] / np.timedelta64(1, 'D')
logins['lifetime'] = logins['lifetime'].astype('int')
logins
```

```python
# Откинем время из даты

logins['event_timestamp'] = logins['event_timestamp'].dt.to_period('D').dt.start_time
logins
```

```python
# Получим таблицу, в которой у каждой когорты на каждый день будет отображаться количество пользователей, зашедших в игру

logins_cohort = logins.groupby(['week_start','lifetime'], as_index=False).agg({'client_id':'nunique'})
logins_cohort
```

```python
# Получим размер каждой когорты

installs_cohort = logins.groupby(['week_start'], as_index=False).agg({'client_id':'nunique'}
                                                                      ).rename(columns = {'client_id':'users_count'})
installs_cohort
```

```python
# Теперь смержим две предыдущие таблицы.
# Теперь имеем каждый день для каждой когорты, количество пользователей, зашедших в игру, и размер когорты

logins_cohort = logins_cohort.merge(
    installs_cohort,
    on='week_start',
    how='inner'
)

# И косметика
logins_cohort = logins_cohort.rename(columns={'client_id':'users_count', 'users_count':'cohort_size'})
logins_cohort
```

```python
# Теперь получаем непосредственно retention и косметика

logins_cohort['retention, %'] = (logins_cohort['users_count'] / logins_cohort['cohort_size'] * 100).round(2)
logins_cohort.sort_values(by='lifetime').reset_index().drop('index', axis=1)
```
